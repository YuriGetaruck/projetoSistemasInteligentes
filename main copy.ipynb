```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
data =  open("sentencas_sem_duplicatas.txt", encoding="utf-8")
meuTexto = data.read()
```

```python
mytokenizer = Tokenizer()
mytokenizer.fit_on_texts([meuTexto])
total_words = len(mytokenizer.word_index) + 1
```

```python
mytokenizer.word_index
```

```python
my_input_sequences = []
for line in meuTexto.split('\n'):
    #print(line)
    token_list = mytokenizer.texts_to_sequences([line])[0]
    #print(token_list)
    for i in range(1, len(token_list)):
        my_n_gram_sequence = token_list[:i+1]
        #print(my_n_gram_sequence)
        my_input_sequences.append(my_n_gram_sequence)
        #print(input_sequences)
```

```python
max_sequence_len = max([len(seq) for seq in my_input_sequences])
input_sequences = np.array(pad_sequences(my_input_sequences, maxlen=max_sequence_len, padding='pre'))
```

```python
input_sequences[1]
```

```python
X = input_sequences[:, :-1]
y = input_sequences[:, -1]
```

```python
X[1]
```

```python
y
```

```python
y = np.array(tf.keras.utils.to_categorical(y, num_classes=total_words))
```

```python
y[1]
```

```python
model = Sequential()
model.add(Embedding(total_words, 100, input_length=max_sequence_len-1))
model.add(LSTM(150))
model.add(Dense(total_words, activation='softmax'))
print(model.summary())
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=10, verbose=1)
```

```python
model.save('modelo_LSVM2.h5')
```

```python
input_text = "eu gosto"
predict_next_words= 1

for _ in range(predict_next_words):
    token_list = mytokenizer.texts_to_sequences([input_text])[0]
    print(token_list)
    token_list = pad_sequences([token_list], maxlen=max_sequence_len-1, padding='pre')
    predicted = np.argmax(model.predict(token_list), axis=-1)
    output_word = ""
    for word, index in mytokenizer.word_index.items():
        if index == predicted:
            output_word = word
            break
    input_text += " " + output_word

print(input_text)
```

```python
import re

# Abrindo o arquivo e lendo o texto com a codificação 'utf-8'
with open('DBpediaRelations-PT-0.2.txt', 'r', encoding='utf-8') as file:
    texto = file.read()

# Usando regex para encontrar as sentenças únicas
sentences = re.findall(r'SENTENCE : (.*?)\n', texto, re.DOTALL)

# Removendo duplicatas utilizando um conjunto (set)
sentences_unique = set(sentences)

# Juntando as sentenças únicas encontradas em um novo texto
novo_texto = '\n'.join(sentences_unique)

# Escrevendo as sentenças únicas em um novo arquivo
with open('sentencas_sem_duplicatas.txt', 'w', encoding='utf-8') as file:
    file.write(novo_texto)

print("As sentenças únicas foram salvas em 'sentencas_sem_duplicatas.txt'")

```
