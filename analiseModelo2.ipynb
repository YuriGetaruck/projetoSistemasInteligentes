```python
import matplotlib.pyplot as plt

# Dados do histórico (substitua pelos seus dados)
loss_values = [5.1927, 4.0074, 3.5231, 3.1121, 2.7370, 2.4001, 2.1039, 1.8543, 1.6413, 1.4604]
accuracy_values = [0.3162, 0.4068, 0.4362, 0.4627, 0.4913, 0.5292, 0.5711, 0.6120, 0.6499, 0.6842]

# Número de épocas
epochs = range(1, len(loss_values) + 1)

# Plotando a perda ao longo das épocas
plt.plot(epochs, loss_values, 'bo', label='Perda')
plt.title('Perda durante o Treinamento')
plt.xlabel('Épocas')
plt.ylabel('Perda')
plt.legend()
plt.show()

# Plotando a precisão ao longo das épocas
plt.plot(epochs, accuracy_values, 'ro', label='Precisão')
plt.title('Precisão durante o Treinamento')
plt.xlabel('Épocas')
plt.ylabel('Precisão')
plt.legend()
plt.show()
```
